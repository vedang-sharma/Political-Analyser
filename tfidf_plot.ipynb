```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('t_values/TFIDF_mean.csv')
df
```

```python
years = ['1999', '2004', '2009', '2014']
```

```python
names = ['EDUCATION', 'INFRASTRUCTURE', 'AGRICULTURE', 'HEALTH', 'ECONOMICS']
```

```python
for i in range(5):
    plt.plot(years, df.iloc[::2, i] ,color = 'tab:orange', label = 'National Democratic Alliance')
    plt.plot(years, df.iloc[1::2, i], color = 'dodgerblue', label = 'United Progressive Alliance')
    
    plt.legend()
    
    plt.title(names[i], fontsize = 16, fontweight = 'bold')
    plt.xlabel('Years', fontsize = 14)
    plt.ylabel('TF-IDF Values', fontsize = 14)
    
    plt.savefig('t_values/'+'white'+names[i]+'.png', dpi = 300, bbox_inches = 'tight', facecolor = 'snow')
    plt.show()
```
