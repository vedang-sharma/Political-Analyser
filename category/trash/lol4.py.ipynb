```python
import pandas as pd
```

```python
with open('C:/Users/jayes/Desktop/collegeProject/mainefesto/temp/NDA2004.xml', encoding = 'utf-8') as file_obj:
    df = pd.DataFrame(file_obj.readlines(), columns = ['sentences'])
df.reset_index(drop = True)
df.head()
```

```python
import re
corpus = np.array([])

lemmatizer = WordNetLemmatizer()

for i in range(len(df['sentences'])-1):
    review = re.sub('[^a-zA-Z]', ' ', df['sentences'][i+1])
    review = review.lower()
    review = review.split()
    review = [lemmatizer.lemmatize(word) for word in review if word not in set(stopwords.words('English'))] #Lemmatization of each word is done here
    
    corpus = np.append(corpus, ' '.join(review))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
model = TfidfVectorizer()
X = model.fit_transform(corpus)

data = pd.DataFrame(X.toarray(), columns = [model.get_feature_names_out()]) #Converting the 'X' into a Dataframe
data.head()
```

```python
e = []
for i in range(len(edu)):
    edu[i] = re.sub('[^a-zA-Z]', ' ', edu[i]) #Removing unnecessary characters from the word
    edu[i] = lemmatizer.lemmatize(edu[i].lower())
for i in range(len(agro)):
    agro[i] = re.sub('[^a-zA-Z]', ' ', agro[i]) #Removing unnecessary characters from the word
    agro[i] = lemmatizer.lemmatize(agro[i].lower())
for i in range(len(infra)):
    infra[i] = re.sub('[^a-zA-Z]', ' ', infra[i]) #Removing unnecessary characters from the word
    infra[i] = lemmatizer.lemmatize(infra[i].lower())
for i in range(len(hea)):
    hea[i] = re.sub('[^a-zA-Z]', ' ', hea[i]) #Removing unnecessary characters from the word
    hea[i] = lemmatizer.lemmatize(hea[i].lower())
for i in range(len(econo)):
    econo[i] = re.sub('[^a-zA-Z]', ' ', econo[i]) #Removing unnecessary characters from the word
    econo[i] = lemmatizer.lemmatize(econo[i].lower())
```

```python
finale = [] #List to store all the TF - IDF values of a sentence

for i in range(len(corpus)):
    valagri = [] #TF - IDF values are stored for each sentence here
    valeco = []
    valedu = []
    valinfra = []
    valhea = []
    
    for word in corpus[i].split():
        if word in set(edu):
            valedu.append(data[word].iloc[i][0])
        if word in set(agro):
            valagri.append(data[word].iloc[i][0])
        if word in set(econo):
            valeco.append(data[word].iloc[i][0])
        if word in set(infr):
            valinfra.append(data[word].iloc[i][0])
        if word in set(hea):
            valhea.append(data[word].iloc[i][0])        
    
    vedu = sum(valedu)/len(valedu) if len(valedu) else 0
    vagri = sum(valagri)/len(valagri) if len(valagri) else 0
    vecono = sum(valeco)/len(valeco) if len(valeco) else 0
    vhea = sum(valhea)/len(valhea) if len(valhea) else 0
    vinfra = sum(valinfra)/len(valinfra) if len(valinfra) else 0
    
    finale.append([vedu,vagri,vecono,vinfra,vhea,i])
finale
```

```python
edu_df = pd.DataFrame(finale, columns = ['valedu','vagri','vecono','vinfra','vhea','sno'])
edu_df.to_csv('C:/Users/jayes/Desktop/collegeProject/mainefesto/2004_education.csv')
```
